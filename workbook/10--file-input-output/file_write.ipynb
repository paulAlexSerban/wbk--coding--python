```python
import os
```

```python
def write_file(file_path: str, text: str) -> int:
    file = None
    try:
        file = open(file_path, "w")
        return file.write(text)
    except OSError:
        print(f"Could not open file at path {file_path}")
    finally:
        if file != None:
            file.close()


file_name = "myfile.txt"
count = write_file(file_name, "Hello Python")

if count is not None:
    print(f"{count} characters written to file {file_name}")
else:
    print(f"Could not write to file {file_name}")


print(os.path.abspath(os.getcwd()))
```
