```python
def read_file(file_path: str):
    """
    file_path: string
    """
    file = None
    try:
        file = open(file_path)
    except OSError:
        print(f"Error! Could not open the file at path {file_path}")
    else:
        lines = []
        for line in file:
            lines.append(line)
        return lines
    finally:
        if file is not None:
            file.close()
```

```python
text = read_file("myfile.txt")

if text is not None:
    print(text)
```
