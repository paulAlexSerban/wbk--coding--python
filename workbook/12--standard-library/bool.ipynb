Working with the bool() Function
- Definition: If you want to check whether a value is truthy or falsy, you can use the built-in bool() function. 
- It explicitly converts a value to its boolean equivalent and returns True for truthy values and False for falsy values. Here are a few examples:

```python
print(bool(False)) # False
print(bool(0))  # False
print(bool('')) # False

print(bool(True)) # True
print(bool(1)) # True
print(bool('Hello')) # True
```
