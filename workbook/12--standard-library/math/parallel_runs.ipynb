Given COUNT_GAMES and COUNT_PARALLELS with a duration of MINUTES_PER_TEST.
What is the total time to run all tests in batches?

```python
import math

count_games = 106
count_parallels = 10
minutes_per_test = 10

batches = math.ceil(count_games / count_parallels)
total_minutes = batches * minutes_per_test

print(f"count_Games: {count_games}")
print(f"Parallel slots: {count_parallels}")
print(f"Time per test: {minutes_per_test} min")
print(f"Batches needed: {batches}  (ceil({count_games}/{count_parallels}))")
print(f"Total time: {total_minutes} minutes")
```

The key is `math.ceil(29 / 10) = 3` batches - batch 1 runs games 1-10, batch 2 runs 11-20, batch 3 runs the remaining 9 (still takes the full min since you're waiting for the last one to finish).
