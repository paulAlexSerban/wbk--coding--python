```python
import math

games = 106
parallel = 10
minutes_per_test = 10

batches = math.ceil(games / parallel)
total_minutes = batches * minutes_per_test

print(f"Games: {games}")
print(f"Parallel slots: {parallel}")
print(f"Time per test: {minutes_per_test} min")
print(f"Batches needed: {batches}  (ceil({games}/{parallel}))")
print(f"Total time: {total_minutes} minutes")
```
