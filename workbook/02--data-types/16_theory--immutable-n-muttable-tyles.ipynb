# Immutable and Mutable Types
Immutable Types: 
- These types cannot change once declared, although you can point their variables at something new, which is called reassignment. 
- They include integer, float, complex, boolean, string, tuple, range, and None.

Mutable Types: 
- These types can change once declared. You can add, remove, or update their items. 
- They include collection types such as list, set, and dictionary.

type() Function: To see the type for a variable, you can use the type() function like this:

```python
greeting = 'Hello there!'
age = 21

print(type(greeting)) # <class 'str'>
print(type(age)) # <class 'int'>
```

isinstance() Function: This is used to check if a variable matches a specific data type:

```python
greeting = 'Hello world'
name = 'John Doe'

print(isinstance(greeting, str)) # True
print(isinstance(name, int)) # False
```
