```python
# integer variable.
integer_number=100
print("The type of variable having value", integer_number, " is ", type(integer_number))
```

```python
# float variable.
float_number=20.345
print("The type of variable having value", float_number, " is ", type(float_number))
```

```python
# complex variable.
complex_number=10+3j
print("The type of variable having value", complex_number, " is ", type(complex_number))
```
